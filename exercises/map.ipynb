```python
map = folium.Map(tiles="cartodb positron")

ostan_geo = requests.get(
    "https://raw.githubusercontent.com/mrunderline/iran-geojson/master/iran_geo.json"
).json()

ostan_data = pandas.read_csv(
    "../data.csv"
)

map = folium.Map(location=[32, 53], zoom_start=5)

folium.Choropleth(
    geo_data=ostan_geo,
    name="choropleth",
    data=ostan_data,
    columns=["Ostan", "pythonProgrammer"],
    key_on="feature.properties.NAME_1",
    fill_color="YlGn",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="python programmers (%)",
).add_to(map)

folium.GeoJson(ostan_geo, name="hello map").add_to(map)

folium.LayerControl().add_to(map)

map
```
